```python
# To fill the null Values with zeros.
vac.fillna(0,inplace=True)
```

```python
# To View the  top data points after cleaning the Data.
vac.head()
```

```python
# To find the total vaccination of each Country
total_vaccination_data={}
for country in vac.country.unique():
    vaccination=0
    for i in range(len(vac)):
        if vac.country[i]==country:
            vaccination=vaccination+vac.total_vaccinations[i]
    
    total_vaccination_data[country]=vaccination
    
    country_wise_vaccination=pd.DataFrame.from_dict(total_vaccination_data,orient='index',columns=['Country_wise_vaccination_now'])
    
       
```

```python
# Total vaccination of Each country
country_wise_vaccination
```

```python
# Most Vaccinated Countries
country_wise_vaccination.sort_values(by='Country_wise_vaccination_now',ascending=False)[0:5]
```

```python
# Daily vaccination Visualisation in recent trend
plt.figure(figsize=(10,10))
sns.lineplot(x=vac['date'].tail(),y=vac['daily_vaccinations_per_million'],data=vac)
plt.show()
```
